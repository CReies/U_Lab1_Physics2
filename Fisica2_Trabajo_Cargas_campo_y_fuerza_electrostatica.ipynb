####Problema 1. Fuerza entre dos partículas cargadas $\color{green}{\text{(4.5 puntos)}}$
---

Una partícula con una carga de $5~\mu C$ se ubica en el plano cartesiano, en el punto $P(-2, 1)$. Una segunda partícula, con una carga de $-8~\mu C$, se ubica en el punto
$Q(1, 5)$. La distancia, tanto en el eje vertical como en el horizontal, se mide en centímetros.

a) 	Calcula la magnitud de la fuerza ejercida por la primera partícula sobre la segunda. ¿La fuerza es de atracción o de repulsión? Explica tu respuesta.

>**Solución:**

Primero hay que calcular la distancia en $m$ que hay entre las cargas

$
d=\sqrt{(y_2-y_1)^2+(x_2-x_1)^2}\\
=\sqrt{(5-1)^2+(1-(-2))^2}\qquad (cm)\\
=\sqrt{4^2+3^2}
=\sqrt{16+9}
=\sqrt{25}\\
=\boxed{5cm}
=\frac{5}{100}m
=\boxed{\frac{1}{20}m}
$

Luego hay que convertir los $\mu C$ en $C$

$
1\mu C = 1\times10^{-6}C\\
A=5\mu C = \boxed{5\times10^{-6}C}\\
B=-8\mu C = \boxed{-8\times10^{-6}C}\\
$

Ya con estos datos se puede calcular la fuerza electromagnética aplicando su formula

$
F_e=K\big(\frac{q_1\times q_2}{r^2}\big)\\
=K\big(\frac{(5\times10^{-6})(-8\times10^{-6})}{(\frac{1}{20})^2}\big)\qquad (\frac{Nm^2}{C^2})(\frac{C^2}{m^2})\\
=K\big(\frac{-40\times10^{-12}}{\frac{1}{400}}\big)\qquad (N)\\
=K\big(\frac{-4\times10^{-11}}{\frac{1}{400}}\big)
=K((-4\times10^{-11})(400))
=K(-1.6\times10^{-8})\\
\approx9\times10^{9}(-1.6\times10^{-8})\\
=-14.4\times10^1
=\boxed{-144N}
$

Con esto ahora se sabe que hay una fuerza de aproximadamente $144N$, tambien se sabe que es una fuerza de atraccion debido a que la formula dio un valor negativo

```python
import math


def emForce(q1, q2, xy1, xy2):
  d = abs(math.sqrt((xy2[0] - xy1[0]) ** 2 + (xy2[1] - xy1[1]) ** 2)/100)
  k = 9 * (10**9)
  return k * ((q1 * q2) / (d**2))

q1 = float(input("Ingrese el valor de la primera carga: "))
q2 = float(input("Ingrese el valor de la segunda carga: "))
xy1 = input("Ingrese las coordenadas de la primera carga separadas por una coma: x,y: ")
xy2 = input("Ingrese las coordenadas de la segunda carga separadas por una coma: x,y: ")
xy1 = xy1.split(",")
xy2 = xy2.split(",")
xy1 = (float(xy1[0]), float(xy1[1]))
xy2 = (float(xy2[0]), float(xy2[1]))
print("La fuerza entre las 2 cargas es de aproximadamente ", "{:.2e}".format(emForce(q1, q2, xy1, xy2)),"N",)
```
